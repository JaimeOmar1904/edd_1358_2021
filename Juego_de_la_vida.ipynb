```
class Array2D: #Clase Array2D
	def __init__(self,ren,col):
		self.renglon = ren
		self.columna = col
		self.arreglo =[ [0 for i in range(col)] for i in range(ren)]

	def clear(self,dato):
		self.arreglo =[ [[dato]for i in range(self.columna)] for i in range(self.renglon)]

#Metodo de get y set

	def get_row_size(self):# Regresa el Tamaño de los renglones
		return self.renglon
	def get_col_size(self):# Regresa el tamaño de las columnas
		return self.columna
	def set_item(self,ren,col,dato):# Regresa el valor de la posicion ren, col.
		self.arreglo[ren][col] = dato
	def get_item(self,ren,col):
		return self.arreglo[ren][col]

#Metodo to_string

	def to_string(self):# Regresa el un objeto string que representa el estado actual
		return self.arreglo
```
